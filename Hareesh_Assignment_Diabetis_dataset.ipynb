```
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```
from google.colab import files
dataset = files.upload()
```

```
import io
df2 = pd.read_csv(io.BytesIO(dataset['datasets_diabetes.csv']))
```

```
df2.head(10)
```

```
x = df2.drop(['Outcome'],axis =1)
y= df2['Outcome']
```

```
x
```

```
from sklearn.model_selection import train_test_split
```

```
x_train,x_test, y_train , y_test = train_test_split(x,y,test_size=0.50 ,random_state=50)
```

```
from keras.utils import np_utils
y_train = np_utils.to_categorical(y_train)
```

```
model=Sequential()
model.add(Dense(20,activation='relu', input_dim=8))
model.add(Dense(10,activation='relu'))
model.add(Dense(2,activation='sigmoid'))
```

```
model.compile(optimizer ='adam',
              loss='binary_crossentropy' ,
              metrics = ['accuracy'])
```

```
model.fit(x_train ,y_train , epochs= 300, batch_size = 50 )
```

```
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
y_pred = np.argmax(y_pred, axis =1 )
accuracy_score(y_test,y_pred)
```
